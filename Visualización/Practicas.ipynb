```r
library(ggplot2)
# Extraemos los datos del paquete ggplot2 llamados mpg
data(mpg, package="ggplot2")

# De los datos, seleccionamos solo aquellos cuyo nombre de la empresa sea
# audi, ford, honda o hyundai
mpg_select <- mpg[mpg$manufacturer %in% c("audi", "ford", "honda", "hyundai"), ]

# Cambiamos el tema default, por uno que nosotros deseemos
theme_set(theme_bw())

# Creamos un objeto ggplot con el data frame que filtramos, y pensamos poner en el eje "x"
# displ y en el "y" la ciudad
# Adems, agregamos ttulo al  plot (title), y le ponemos un pie de imagen(caption)
g <- ggplot(mpg_select, aes(displ, cty)) +
  labs(title="Desplazamiento vs consumo en ciudad", caption="Fuente: mpg
dataset")

# Ahora agregamos los puntos de forma un tanto dispersa (geom_jitter), en el que las columnas
# estarn definidas por la empresa manufactura y el tamao por las highway miles per galon.
# Finalmente, agregamos una lnea de regresin lineal para cada empresa
g + geom_jitter(aes(col=manufacturer, size=hwy)) +
  geom_smooth(aes(col=manufacturer), method="lm", se=F)
```
